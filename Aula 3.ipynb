##EXERCÍCIO 13 - Instrução IF

```python
people = 20
cats = 30
dogs = 15

if people < cats: #Condição verdadeira
  print("Too many cats! The world is doomed!") #Código que será executado se a condição for verdadeira

if people > cats:
  print("Not many cats! The world is saved!")

if people < dogs:
  print("The world is drooled on!")

if people > dogs:
  print("The world is dry!")

dogs += 5

if people >= dogs:
  print("People are greater than or equal to dogs.")

if people <= dogs:
  print("People are less than or equal to dogs.")

if people == dogs:
  print("People are dogs.")
```

###Novidades neste exercício

*   Utilização de estrutura de controle condicional (if) que executa o bloco de código dele somente se a condição é verdadeira.

##EXERCÍCIO 14 - Else e IF

```python
people = 30
cars = 40
trucks = 15

if cars > people:
  print("We should take the cars.")
elif cars < people:
   print("We should not take the cars.")
else:
   print ("we can´t decide.")
if trucks > cars:
   print("That's too many trucks")
elif trucks < cars:
   print("maybe we could take the trucks")
else:
   print("we still can't decide.")

if people > trucks:
   print("Alright, let's just take the trucks.")
else:
   print("Fine, let's stay home then")
```

###Novidades neste exercício

*   Utilização do elif que determina uma nova condição que é testada e executada se a condição do if for falsa e a do elif for verdadeira;
*   Utilização do else que excuta o bloco de código caso as condições do if e o elif sejam falsas.
